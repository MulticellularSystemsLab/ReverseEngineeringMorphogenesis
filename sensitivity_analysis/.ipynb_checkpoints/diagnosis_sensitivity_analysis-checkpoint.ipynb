### 1. Sensitivity analysis for apical and basal frechet errors

```python
# Loading frechet error for upper and lower parameter bounds
error_target_sampled_apical = np.load('error_target_sampled_apical.npy')
error_target_sampled_basal = np.load('error_target_sampled_basal.npy')

# Calculating sensitivity in apical frechet error

sens_frechet_apical = np.zeros((22,))
sens_frechet_basal = np.zeros((22,))

for i in range(22):
    sens_frechet_apical[i] = (error_target_sampled_apical[2*i + 1] - error_target_sampled_apical[2*i]) / 1.4
    sens_frechet_basal[i] = (error_target_sampled_basal[2*i + 1] - error_target_sampled_basal[2*i]) / 1.4
    
sens_frechet_apical_log = np.log(np.abs(sens_frechet_apical))
sens_frechet_basal_log = np.log(np.abs(sens_frechet_basal))

params_lhs = ('T_squ_b','T_cub_b','T_col_b','k_squ_a','k_squ_b','k_squ_l','k_cub_a','k_cub_b','k_cub_l','k_col_a','k_col_b','k_col_l','lo_squ_a','lo_squ_b','lo_squ_l','lo_cub_a','lo_cub_b','lo_cub_l','lo_col_a','lo_col_b','lo_col_l','K_ECM')
y_pos = np.arange(len(params_lhs))

from matplotlib.pyplot import figure
figure(figsize=(8, 4), dpi=200)

plt.subplot(2, 1, 1)
plt.bar(y_pos, sens_frechet_apical_log)
#plt.xticks(y_pos, params_lhs)
plt.xticks([])
plt.ylabel('$log((1/P_{0})(dF_{apical}/dP))$', fontsize=8)

plt.subplot(2, 1, 2)
plt.bar(y_pos, sens_frechet_basal_log)
plt.xticks(y_pos, params_lhs)
plt.xticks(rotation = 90)
plt.ylabel('$log((1/P_{0})(dF_{basal}/dP))$',fontsize=8)
plt.show()

```

### 2. Sensitivity analysis for curvature

```python
curvature_basal_master = np.load('curvature_basal_master.npy')

# Calculating sensitivity in apical frechet error
sens_curv = np.zeros((129,22))

for i in range(22):
    sens_curv[:,i] = np.subtract(curvature_basal_master[2*i+1,:], curvature_basal_master[2*i,:]) / 1.4

sens_curv_log = np.log(np.abs(sens_curv))
    
curv_labels = []
for i in range(129):
    curv_labels.append("C_" + str(i))
    
params_lhs = ('T_squ_b','T_cub_b','T_col_b','k_squ_a','k_squ_b','k_squ_l','k_cub_a','k_cub_b','k_cub_l','k_col_a','k_col_b','k_col_l','lo_squ_a','lo_squ_b','lo_squ_l','lo_cub_a','lo_cub_b','lo_cub_l','lo_col_a','lo_col_b','lo_col_l','K_ECM')
   
import seaborn as sns

sns.set(rc={"figure.dpi":500, 'savefig.dpi':300})
sns.set(rc={'figure.figsize':(4,12)})
ax = sns.heatmap(sens_curv_log, xticklabels = params_lhs, yticklabels = curv_labels, linewidths=.3)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 9)
ax.set_yticklabels(ax.get_ymajorticklabels(), fontsize = 6)
```

```python
fe_exp = FeatureExtractor('input_data/vertices_target_SE.txt', 'log_edges.xlsx')
curv_target = fe_exp.tissue_local_curvature()
index_curv = np.linspace(0, 128, num=129)

fig, ax = plt.subplots()
fig.set_size_inches(6, 6)

for i in range(22):
    plt.subplot(5, 5, i+1)
    plt.plot(index_curv, curv_target,color="green",linewidth=1)
    plt.plot(index_curv, curvature_basal_master[2*i+1,:],color="blue",linewidth=1)
    plt.plot(index_curv, curvature_basal_master[2*i,:],color="red",linewidth=1)
    plt.xticks([])
    plt.yticks([])
    plt.title(params_lhs[i], fontsize=12 )
    plt.grid(color='0.95')
    
plt.subplots_adjust(hspace=0.4)   
plt.show()
```

### 3. Sensitivity analysis of cell height

```python
tissue_edge_length_master = np.load('tissue_edge_length_master.npy')

# Calculating sensitivity in apical frechet error
sens_cell_length = np.zeros((390,22))

for i in range(22):
    sens_cell_length[:,i] = np.subtract(tissue_edge_length_master[2*i+1,:], tissue_edge_length_master[2*i,:]) / 1.4

sens_cell_length_log = np.log(np.abs(sens_cell_length))
    
cell_length_labels = []
for i in range(390):
    cell_length_labels.append("L_" + str(i))
    
params_lhs = ('T_squ_b','T_cub_b','T_col_b','k_squ_a','k_squ_b','k_squ_l','k_cub_a','k_cub_b','k_cub_l','k_col_a','k_col_b','k_col_l','lo_squ_a','lo_squ_b','lo_squ_l','lo_cub_a','lo_cub_b','lo_cub_l','lo_col_a','lo_col_b','lo_col_l','K_ECM')
   
import seaborn as sns
sns.set(rc={"figure.dpi":200, 'savefig.dpi':200})
sns.set(rc={'figure.figsize':(5,32)})
ax = sns.heatmap(sens_cell_length_log, xticklabels = params_lhs, linewidths=.3)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 9)
ax.set_yticklabels(ax.get_ymajorticklabels(), fontsize = 6)
```

```python
fe_exp = FeatureExtractor('input_data/vertices_target_SE.txt', 'log_edges.xlsx')
cell_length_target = fe_exp.edge_length()
index_cell_length = np.linspace(0, 389, num=390)

fig, ax = plt.subplots()
fig.set_size_inches(6, 6)

for i in range(22):
    plt.subplot(5, 5, i+1)
    plt.plot(index_cell_length, cell_length_target,color="green",linewidth=1)
    plt.plot(index_cell_length, tissue_edge_length_master[2*i+1,:],color="blue",linewidth=1)
    plt.plot(index_cell_length, tissue_edge_length_master[2*i,:],color="red",linewidth=1)
    plt.xticks([])
    plt.yticks([])
    plt.title(params_lhs[i], fontsize=12 )
    plt.grid(color='0.95')
    
plt.subplots_adjust(hspace=0.4)   
plt.show()
```

### 4. Correlation analysis of output features

```python
sens_frechet_master = np.zeros((22,2))
sens_frechet_master[:,0] = sens_frechet_apical_log
sens_frechet_master[:,1] = sens_frechet_basal_log
frechet_label = ('F_A','F_B')

df_sens_curv = pd.DataFrame(np.transpose(sens_curv_log), columns = curv_labels)
df_sens_frechet = pd.DataFrame(sens_frechet_master, columns = frechet_label)
df_sens_cell_height = pd.DataFrame(np.transpose(sens_cell_length_log), columns = cell_length_labels)

df_sens_frechet_curv = pd.concat([df_sens_frechet, df_sens_curv], axis=1)
df_sens_frechet_height = pd.concat([df_sens_frechet, df_sens_cell_height], axis=1)
df_sens_frechet_curv_height = pd.concat([df_sens_frechet_curv, df_sens_cell_height], axis=1)

snsplot = sns.pairplot(df_sens_frechet)
```

```python
correlations_frechet_curv = df_sens_frechet_curv.corr()
sns.set(rc={"figure.dpi":200, 'savefig.dpi':200})
sns.set(rc={'figure.figsize':(10,10)})
sns.heatmap(correlations_frechet_curv)
plt.show()
```

```python
correlations_frechet_height = df_sens_frechet_height.corr()
sns.set(rc={"figure.dpi":200, 'savefig.dpi':200})
sns.set(rc={'figure.figsize':(10,10)})
sns.heatmap(correlations_frechet_height)
plt.show()
```

```python
correlations_frechet_curv_height = df_sens_frechet_curv_height.corr()
sns.set(rc={"figure.dpi":200, 'savefig.dpi':200})
sns.set(rc={'figure.figsize':(10,10)})
sns.heatmap(correlations_frechet_curv_height)
plt.show()
```
